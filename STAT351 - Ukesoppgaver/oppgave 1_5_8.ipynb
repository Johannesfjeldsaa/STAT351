```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cdf(x):
    if x < -1:
        return 0
    if -1 <= x and x <1:
        return (x+2) / 4
    if 1 <= x:
        return 1
```

```python
x_values = np.linspace(-2, 4, 10000)
y_values = [cdf(x) for x in x_values]
```

```python
fig, ax = plt.subplots()

ax.grid()
ax.plot(x_values, y_values, label='cdf(x)')
ax.vlines(x=[-1/2, 1/2], ymin=0, ymax=1, colors='red', linestyles= 'dashed', label='a)')
ax.vlines(x=[0], ymin=0, ymax=1, colors='green', linestyles= 'dashed', label='b)')
ax.vlines(x=[1], ymin=0, ymax=1, colors='yellow', linestyles= 'dashed', label='c)')
ax.vlines(x=[2, 3], ymin=0, ymax=1, colors='brown', linestyles= 'dashed', label='d)')

ax.legend(loc='upper left')
```
